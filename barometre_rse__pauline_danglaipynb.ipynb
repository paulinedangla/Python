## I) Phase préparatoire
### A) Importation des données et création des dataframes

```python
#2021
df_2021_part1= pd.read_csv('https://raw.githubusercontent.com/LanCaphe/barometre-rse-vendredi/main/data/2021-partie1.csv')
df_2021_part2= pd.read_csv('https://raw.githubusercontent.com/LanCaphe/barometre-rse-vendredi/main/data/2021-partie2.csv')
df_2021_part3= pd.read_csv('https://raw.githubusercontent.com/LanCaphe/barometre-rse-vendredi/main/data/2021-partie3.csv')
#2022
df_2022_part1= pd.read_csv('https://raw.githubusercontent.com/LanCaphe/barometre-rse-vendredi/main/data/2022-partie1.csv')
df_2022_part2= pd.read_csv('https://raw.githubusercontent.com/LanCaphe/barometre-rse-vendredi/main/data/2022-partie2.csv')
df_2022_part3= pd.read_csv('https://raw.githubusercontent.com/LanCaphe/barometre-rse-vendredi/main/data/2022-partie3.csv')
```

```python
df_2022_part1.head()
```

Merge des 3 parties en un seul dataframe


```python
#rename column #  to Token:
def rename_col_to_token(df):
  df.rename(columns={'#': 'Token'}, inplace=True)

df_list= [df_2021_part2, df_2021_part3,df_2022_part1, df_2022_part2, df_2022_part3]

for i in df_list:
  rename_col_to_token(i)

```

```python
#change column name email (df_2021_part1)
df_2021_part1.rename(columns={'Quelle est votre *adresse email professionnelle* ?': 'email'}, inplace=True)
df_2021_part1.head()
#change column name email [df_2022_part1]
df_2022_part1.rename(columns={'Quelle est votre adresse mail ?': 'email'}, inplace=True)
df_2022_part1.head()
```

```python
#df_2021
df_2021 = df_2021_part1.merge(df_2021_part2, how='right', on='email')

```

```python
df_2021 = df_2021.merge(df_2021_part3, how='right', on='email')
```

```python
#df 2022

df_2022 = df_2022_part1.merge(df_2022_part2, how='right', on="email")
df_2022 = df_2022.merge(df_2022_part3, how='right', on='email')
```
